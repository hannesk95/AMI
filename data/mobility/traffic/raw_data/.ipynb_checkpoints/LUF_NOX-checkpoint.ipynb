```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#change time 24:00 to 00:00 of next day
def my_to_datetime(date_str):
    if date_str[11:13] != '24':
        return pd.to_datetime(date_str, format='%d.%m.%Y %H:%M')

    else:
        date_str = date_str.replace('24', '00')
        return pd.to_datetime(date_str, format='%d.%m.%Y %H:%M') + dt.timedelta(days=1)
```

```python
file = '/Volumes/SD/Messwerte/LFU/NO2_2020_06.xls'
df = pd.read_excel(file, skiprows=2, na_filter=False)

#change time 24:00 to 00:00 of next day
row, col = df.shape
for r in range(0,row):
    df.Zeitpunkt[r] = my_to_datetime(df.Zeitpunkt[r])

df.Zeitpunkt = pd.to_datetime(df.Zeitpunkt).dt.to_period('d')
df = df.groupby('Zeitpunkt').mean()



df = df[['Nürnberg/Bahnhof', 'Nürnberg/Muggenhof', 'Nürnberg/Von-der-Tann-Straße']]
#print(df.head())
df['Nürnberg_mean'] = df.mean(axis=1)
for column in df:
    print(column)
```

```python
path = 'MobilityTrends/TomTom/CitiesGermanyTomTom_daily2020.csv'
df_tt = pd.read_csv(path)
df_tt = df_tt[['Time','traffic_congestion_nuremberg']]
df_tt = df_tt[df_tt['Time'] >= '2020-06-01']
df_tt = df_tt.groupby('Time').mean()
#print(df_tt.head())
```

```python
plt.style.use("ggplot")
ax = df_tt["traffic_congestion_nuremberg"].plot(color="Blue")
plt.title("Nurnberg congestion TomTom")
plt.legend(loc=2)
```

```python
#daily plot
plt.style.use("ggplot")
ax = df["Nürnberg_mean"].plot(color="Green")
plt.title("Nürnberg mean NO2")
plt.legend(loc=2)
#plt.box(False)
```
