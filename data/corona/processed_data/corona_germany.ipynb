```python
import datetime
import pandas as pd
import numpy as np

todays_date = datetime.datetime.now().date()
index_pre_corona = pd.date_range(todays_date-datetime.timedelta(365*30+30*6), periods=30*12, freq='M')

columns_pre_corona = ['cases','deaths']
df_ = pd.DataFrame(index=index_pre_corona, columns=columns_pre_corona)
df_ = df_.fillna(0) # with 0s rather than NaNs
data_pre_corona = np.array([np.zeros(30*12)]*2).T
df_pre_corona = pd.DataFrame(data_pre_corona, index=index_pre_corona, columns=columns_pre_corona)
df_pre_corona['myDt']=index_pre_corona
df_pre_corona['dateRep']=index_pre_corona
```

```python
df = pd.read_csv('corona.csv') # File path description
df.size
df.head()
```

```python
df=df.drop(['popData2019','geoId'], axis=1)
df.dtypes
```

```python
grouped_germany = df[df.countriesAndTerritories == "Germany"]
```

```python
grouped_germany=grouped_germany.drop(['countriesAndTerritories','countryterritoryCode','continentExp'], axis=1)
grouped_germany.head()
```

```python
#grouped_germany.dtypes
#grouped_germany['Date'] = pd.to_datetime(grouped_germany['dateRep'])
#grouped_germany['DOB1'] = grouped_germany['Date'].dt.strftime('%m-%d-%Y')
grouped_germany.dateRep=pd.to_datetime(grouped_germany.dateRep)
grouped_germany.dtypes
```

```python
grouped_germany.shape
```

```python
grouped_germany['myDt']=pd.to_datetime(grouped_germany[['year','month','day']])
grouped_germany=grouped_germany.drop(['year','month','day'], axis=1)
#grouped_germany = grouped_germany.set_index('myDt')
grouped_germany.head()
```

```python
frames= [ grouped_germany,df_pre_corona]
grouped_germany_total = pd.concat(frames,sort=False)
grouped_germany_total.shape
grouped_germany_total.head()
print(grouped_germany_total)
#grouped_germany_total.tail()
```

```python
per_germany = grouped_germany_total.myDt.dt.to_period("M") 
per_germany.head()
#print(per_germany)
per_germany.tail()
```

```python
grouped_germany_monthly = grouped_germany_total.groupby(per_germany)
grouped_germany_monthly.sum()
```

```python
Germany_sum=grouped_germany_monthly.sum()
Germany_sum.head()
print(Germany_sum)
```
