```python
import datetime
import pandas as pd
import numpy as np

todays_date = datetime.datetime.now().date()
index_pre_corona = pd.date_range(todays_date-datetime.timedelta(365*30+30*6), periods=30*12, freq='M')

columns_pre_corona = ['cases','deaths']
df_ = pd.DataFrame(index=index_pre_corona, columns=columns_pre_corona)
df_ = df_.fillna(0) # with 0s rather than NaNs
data_pre_corona = np.array([np.zeros(30*12)]*2).T
df_pre_corona = pd.DataFrame(data_pre_corona, index=index_pre_corona, columns=columns_pre_corona)
df_pre_corona['myDt']=index_pre_corona
df_pre_corona['dateRep']=index_pre_corona
```
