```python
# Installing the packages
import numpy as np
import pandas as pd
import seaborn as sns 
import datetime
import os
```

```python
os.chdir("C:/Uni/Master/AMI/Git/group10/data/economy/processed_data")
test  = pd.read_csv('UmsatzGastgewerbe.csv',encoding = "ISO-8859-1", sep=';')

test.head()
#test.dtypes
```

```python
#test=test.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
test.columns = ['buisnesstype','date', 'month', 'Originalwerte']
#pd.options.display.float_format = '{:,.0f}'.format
test= test.dropna()
test['date'] = test['date'].apply(np.int64)
```

```python
test=test.replace(to_replace=r'Januar', value='01', regex=True)
test=test.replace(to_replace=r'Februar', value='02', regex=True)
test=test.replace(to_replace=r'März', value='03', regex=True)
test=test.replace(to_replace=r'April', value='04', regex=True)
test=test.replace(to_replace=r'Mai', value='05', regex=True)
test=test.replace(to_replace=r'Juni', value='06', regex=True)
test=test.replace(to_replace=r'Juli', value='07', regex=True)
test=test.replace(to_replace=r'August', value='08', regex=True)
test=test.replace(to_replace=r'September', value='09', regex=True)
test=test.replace(to_replace=r'Oktober', value='10', regex=True)
test=test.replace(to_replace=r'November', value='11', regex=True)
test=test.replace(to_replace=r'Dezember', value='12', regex=True)
test.head()
```

```python
test.date= test['date'].astype(str)+'-'+test['month']
test=test.drop(['month'], axis=1)
test.head()
```

```python
test = test[test['Originalwerte'] != '...']
test=test.replace(to_replace=r'Beherbergung', value='ECO_Accommodation', regex=True)
test=test.replace(to_replace=r'Hotels, Gasthöfe und Pensionen', value='ECO_Hotels_Inns_Guesthouses', regex=True)
test=test.replace(to_replace=r'Gastronomie', value='ECO_Gastronomy', regex=True)
test=test.replace(to_replace=r',', value='.', regex=True)
```

```python
grouped_Beherbergung = test[test.buisnesstype == "ECO_Accommodation"]
grouped_Hotels = test[test.buisnesstype == "ECO_Hotels_Inns_Guesthouses"]
grouped_Gastronomie = test[test.buisnesstype == "ECO_Gastronomy"]
grouped_Hotels.head()
#grouped_Beherbergung.head()
#grouped_Gastronomie.hea()
```

```python
grouped_Beherbergung=grouped_Beherbergung.drop(['buisnesstype'], axis=1)
grouped_Beherbergung.columns = ['date', 'ECO_Accommodation']

grouped_Hotels=grouped_Hotels.drop(['buisnesstype'], axis=1)
grouped_Hotels.columns = ['date', 'ECO_Hotels_Inns_Guesthouses']

grouped_Gastronomie=grouped_Gastronomie.drop(['buisnesstype'], axis=1)
grouped_Gastronomie.columns = ['date', 'ECO_Gastronomy']
```

```python
grouped_Beherbergung.to_csv(r'../ECO_Accommodation.csv', index = False)
grouped_Hotels.to_csv(r'../ECO_Hotels_Inns_Guesthouses.csv', index = False)
grouped_Gastronomie.to_csv(r'../ECO_Gastronomy.csv', index = False)
```
