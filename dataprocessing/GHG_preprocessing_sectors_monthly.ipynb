### Create dataframe in predefined style and save as .csv into file

```python
d = {'date': years, 'E_Mio.tonnes_CO2': np.asarray(energiewirtschaft_1990_2017) + np.asarray(haushalte_1990_2017), 'M_Mio.tonnes_CO2': verkeher_1990_2017, 'ECO_Mio.tonnes_CO2': np.asarray(industrie_1990_2017) + np.asarray(gewerbe_1990_2017)}
df = pd.DataFrame(data=d)
df
#df.to_csv('./Oeko-Institut_Sektorale_Abgrenzung_Treibhausgasemissionen_Daten_Sektor.csv', index=False)
```

```python
valuesE = df['E_Mio.tonnes_CO2']
valuesM = df['M_Mio.tonnes_CO2']
valuesECO = df['ECO_Mio.tonnes_CO2']
time = np.arange(1990, 2021, 1)
```

```python
sns.lineplot(time, valuesE.tolist())
sns.lineplot(time, valuesM.tolist())
sns.lineplot(time, valuesECO.tolist())
sns.lineplot(time, valuesE+valuesM+valuesECO)
plt.title('CO2 Emissions 1990-2020')
plt.xlabel('Years')
plt.ylabel('CO2 in Mio. Tonnes')
plt.ylim(0, 1400)
plt.xticks(years, rotation=90)
plt.grid()
```

```python
df_adopt = pd.read_csv('../data/greenhouse_emissions/raw data/TNO-MACCco/timeprofiles-month-in-year_GNFR.csv', skiprows=6, delimiter=';')
```

```python
df_adopt.index = df_adopt['GNFR_Category_Name']
df_adopt = df_adopt.drop(['Unnamed: 0', 'TNO GNFR sectors Sept 2018', 'GNFR_Category_Name'], axis=1)
df_adopt['sum'] = df_adopt.sum(axis=1)
```

```python
df_adopt
```

```python
E = []

for i in range(len(valuesE)):
    E.extend(((valuesE[i] / 12) * df_adopt.iloc[0, 0:-1]).tolist())
```

```python
M = []

for i in range(len(valuesM)):
    M.extend((valuesM[i] / 12) * df_adopt.iloc[5, 0:-1])
```

```python
ECO = []

for i in range(len(valuesECO)):
    ECO.extend((valuesECO[i] / 12) * df_adopt.iloc[1, 0:-1])
    
len(ECO)
```

```python
sns.lineplot(np.arange(0, 372, 1), E)
plt.title('CO2 Emissions 1990-2020 - Energy & Households monthly')
```

```python
sns.lineplot(np.arange(0, 372, 1), M)
plt.title('CO2 Emissions 1990-2020 - Mobility monthly')
```

```python
sns.lineplot(np.arange(0, 372, 1), ECO)
plt.title('CO2 Emissions 1990-2020 - Economy monthly')
```

```python
from datetime import datetime, timedelta
from collections import OrderedDict
dates = ["1990-01-01", "2020-12-31"]
start, end = [datetime.strptime(_, "%Y-%m-%d") for _ in dates]
monthly_dates = list(OrderedDict(((start + timedelta(_)).strftime(r"%b-%y"), None) for _ in range((end - start).days)).keys())
```

```python
d = {'date': monthly_dates, 'E_Mio.tonnes_CO2': E, 'M_Mio.tonnes_CO2': M, 'ECO_Mio.tonnes_CO2': ECO}
df = pd.DataFrame(data=d)
#df.to_csv('./Oeko-Institut_Sektorale_Abgrenzung_Treibhausgasemissionen_Daten_Sektor.csv', index=False)
```

```python
df
```

```python
df['Total_CO2_Emission'] = df.sum(axis=1)
df.index = df['date']
```

```python
df = df.drop(['date'], axis=1)
```
