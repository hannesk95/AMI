```python
os.chdir("C:/Uni/Master/AMI/Git/group10/data/economy/processed_data")
test  = pd.read_csv('Number_Unemployment.csv',encoding = "ISO-8859-1", sep=';')
test.head()
```

```python
test.dtypes
```

```python
test.columns = ['area', 'year', 'month', 'Arbeitslose', 'Arbeitslose_quote_allerZivil', 'Arbeitslose_quote_abhZivil', 'gemeldeteArbeitsstellen', 'Kurzarbeiter', 'Kurzarbeitsbetriebe']
test.head()
```

```python
test.dtypes
```

```python
test= test.dropna()
test['year']=test['year'].astype(int)
#test['year'].astype(int)
```

```python
test=test.replace(to_replace=r'Januar', value='01', regex=True)
test=test.replace(to_replace=r'Februar', value='02', regex=True)
test=test.replace(to_replace=r'März', value='03', regex=True)
test=test.replace(to_replace=r'April', value='04', regex=True)
test=test.replace(to_replace=r'Mai', value='05', regex=True)
test=test.replace(to_replace=r'Juni', value='06', regex=True)
test=test.replace(to_replace=r'Juli', value='07', regex=True)
test=test.replace(to_replace=r'August', value='08', regex=True)
test=test.replace(to_replace=r'September', value='09', regex=True)
test=test.replace(to_replace=r'Oktober', value='10', regex=True)
test=test.replace(to_replace=r'November', value='11', regex=True)
test=test.replace(to_replace=r'Dezember', value='12', regex=True)
test.head()
```

```python
test.year= test['year'].astype(str) + '-' + test['month']
test.head()
```

```python
#only keep area= insgesamt
grouped = test[test.area == "Insgesamt"]
grouped.head()
```

```python
grouped.columns = ['type', 'date', 'month', 'ECO_Unemployment_Number', 'Arbeitslose_quote_allerZivil', 'Arbeitslose_quote_abhZivil', 'gemeldeteArbeitsstellen', 'Kurzarbeiter', 'Kurzarbeitsbetriebe']
grouped.head()
```

```python

grouped = grouped.loc[grouped['type'] == 'Insgesamt']
```

```python
#grouped.date= grouped['date'].astype(str)+'-'+grouped['month']
grouped=grouped.drop(['type'], axis=1)
grouped.head()
```

```python
grouped_out = grouped[['date','ECO_Unemployment_Number']]
print(grouped_out)
```
